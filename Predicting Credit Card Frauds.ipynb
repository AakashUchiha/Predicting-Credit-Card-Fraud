```python
from sklearn.ensemble import IsolationForest
```

```python
iso_for=IsolationForest(n_estimators=100,max_samples=len(X),random_state=42,contamination=outlier_fraction)
```

```python
y_pred[y_pred==1]=0
y_pred[y_pred==-1]=1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
